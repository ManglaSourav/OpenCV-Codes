## Accessing and Manipulating Pixels

```python
(b, g, r) = image[50,50]
print("Pixel at (50, 50) - Red: {} Green: {}, Blue: {}".format(r,g,b))
```

```python
image[50, 50] = (0, 0, 0)
(b, g, r) = image[0, 0]
print("Pixel at (50, 50) - Red: {}, Green: {}, Blue: {}".format(r,
g, b))
```

OpenCV stores RGB
channels in reverse order. While we normally think in terms
of Red, Green, and Blue, OpenCV actually stores them in
the order of Blue, Green, and Red.

## Line

```python
canvas = np.zeros((300, 300, 3), dtype = "uint8")
```

```python
red = (0, 0, 255) #(b,g,r)
green = (0, 255, 0)
blue = (255, 0, 0)
```

```python
#        (image,starting_pt, end_pt, color_of_line,thickness_of_line(-1 for solid fill))
cv2.line(canvas, (0,0), (300,300), green, 3)
plt.imshow(canvas)
```

## rectangle

```python
cv2.rectangle(canvas, (150, 10), (200, 90), green)
plt.imshow(canvas)
```

## circles

```python
canvas = np.zeros((300, 300, 3), dtype = "uint8")
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
white = (255, 255, 255)
```

```python
for r in range(0, 175, 25):
    cv2.circle(canvas, (centerX, centerY), r, white)
```

```python
plt.imshow(canvas)
```

```python
for i in range(0, 25):
    radius = np.random.randint(5, high = 200)
    color = np.random.randint(0, high = 256, size = (3,)).tolist()
    pt = np.random.randint(0, high = 300, size = (2,))
    cv2.circle(canvas, tuple(pt), radius, color, -1)
plt.imshow(canvas)
```
